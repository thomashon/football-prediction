```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df['date'] = df['date'].str[4:]
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'] + ' ' + df['time'])
```

```python
df.head()
```

```python
df['weekday'] = df['date'].dt.weekday
```

```python
df['day'] = df['date'].dt.day
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
df.head()
```

```python
df.index = df['date']
```

```python
df.head()
```

```python
dictionary_home_coach = {}
for i in range(len(df.date)):
    if type(df.home_coach[i]) == str:
        dictionary_home_coach[df.identifier[i]] = df.home_coach[i]

        
dictionary_guest_coach = {}
for i in range(len(df.date)):
    if type(df.guest_coach[i]) == str:
        dictionary_guest_coach[df.identifier[i]] = df.guest_coach[i]

        
dictionary_home_game_system = {}
for i in range(len(df.date)):
    if type(df.home_game_system[i]) == str:
        dictionary_home_game_system[df.identifier[i]] = df.home_game_system[i]


dictionary_guest_game_system = {}
for i in range(len(df.date)):
    if type(df.guest_game_system[i]) == str:
        dictionary_guest_game_system[df.identifier[i]] = df.guest_game_system[i]
```

```python
df.home_coach = df.identifier.map(dictionary_home_coach)
df.guest_coach = df.identifier.map(dictionary_guest_coach)
df.home_game_system = df.identifier.map(dictionary_home_game_system)
df.guest_game_system = df.identifier.map(dictionary_guest_game_system)
```

```python
df.shape
```

```python
df = df.drop_duplicates()
```
